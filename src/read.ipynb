```python
import pandas as pd
import json
import gzip
```

```python
def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield json.loads(l)


def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')
```

```python
meta = getDF('data/meta_Digital_Music.json.gz')
rev = getDF('data/Digital_Music.json.gz')
```
